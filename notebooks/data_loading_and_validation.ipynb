## Objective

To load all raw Olist datasets and validate their structural integrity by checking shapes, data types, missing values, duplicates, and primary key uniqueness.  
No joins, cleaning logic, or visualizations are performed in this notebook.

```python
import numpy as np
import pandas as pd
```

```python
# Load all datasets from the data/original folder
orders = pd.read_csv('data/original/olist_orders_dataset.csv')
customers = pd.read_csv('data/original/olist_customers_dataset.csv')
products = pd.read_csv('data/original/olist_products_dataset.csv')
order_items = pd.read_csv('data/original/olist_order_items_dataset.csv')
order_reviews = pd.read_csv('data/original/olist_order_reviews_dataset.csv')
order_payments = pd.read_csv('data/original/olist_order_payments_dataset.csv')
sellers = pd.read_csv('data/original/olist_sellers_dataset.csv')
product_category_name_translation = pd.read_csv('data/original/product_category_name_translation.csv')

# Display summary of loaded datasets
print("Orders shape:", orders.shape)
print("Customers shape:", customers.shape)
print("Products shape:", products.shape)
print("Order Items shape:", order_items.shape)
print("Order Reviews shape:", order_reviews.shape)
print("Order Payments shape:", order_payments.shape)
print("Sellers shape:", sellers.shape)
print("Product Category Translation shape:", product_category_name_translation.shape)
```
