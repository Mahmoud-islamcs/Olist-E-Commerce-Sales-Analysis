## Objective

To load all raw Olist datasets and validate their structural integrity by checking shapes, data types, missing values, duplicates, and primary key uniqueness.  
No joins, cleaning logic, or visualizations are performed in this notebook.

```python
import numpy as np
import pandas as pd
```

```python
customers = pd.read_csv('olist_customers_dataset.csv')
customers.head()
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
orders.head()
```

```python
items = pd.read_csv("olist_order_items_dataset.csv")
items.head()
```

```python
payments = pd.read_csv("olist_order_payments_dataset.csv")
payments.head()
```

```python
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
reviews.head()
```

```python
geolocations = pd.read_csv("olist_geolocation_dataset.csv")
geolocations.head()
```

```python
sellers = pd.read_csv("olist_sellers_dataset.csv")
sellers.head()
```

```python
products = pd.read_csv("olist_products_dataset.csv")
products.head()
```

```python
name_translations = pd.read_csv("product_category_name_translation.csv")
name_translations.head()
```

```python
# Check shapes
print(f"Customers dataset shape {customers.shape}")                 # 1
print(f"Orders dataset shape {orders.shape}")                       # 2
print(f"Items dataset shape {items.shape}")                         # 3
print(f"Payments dataset shape {payments.shape}")                   # 4
print(f"Reviews dataset shape {reviews.shape}")                     # 5
print(f"Geolocations dataset shape {geolocations.shape}")           # 6
print(f"Sellers dataset shape {sellers.shape}")                     # 7
print(f"Products dataset shape {products.shape}")                   # 8
print(f"Name translations dataset shape {name_translations.shape}") # 9
```

```python
# Check dtypes
dataframes = {
    "Customers": customers,
    "Orders": orders,
    "Items": items,
    "Payments": payments,
    "Reviews": reviews,
    "Geolocations": geolocations,
    "Sellers": sellers,
    "Products": products,
    "Name Translations": name_translations
}

for name, df in dataframes.items():
    print(f"{name} dtypes :")
    print(df.dtypes)
    print("\n")
```

```python
# Missing values summary
for name, df in dataframes.items():
    missing = df.isnull().sum()
    if missing.sum() > 0:
        print(f"Missing Values in {name} :")
        print(missing[missing > 0])
        print("\n")
    else:
        print(f"{name}: No missing values found !!\n")
```

```python
# Duplicate keys check
key_audit = {
    "Customers": [customers, "customer_id"],
    "Orders": [orders, "order_id"],
    "Items": [items, "order_item_id"],
    "Payments": [payments, "order_id"],
    "Reviews": [reviews, "review_id"],
    "Sellers": [sellers, "seller_id"],
    "Products": [products, "product_id"]
}
for name, (df, key) in key_audit.items():
    duplicates = df.duplicated(subset=[key]).sum()
    if duplicates > 0:
        print(f"{name}: Found {duplicates} duplicate keys in '{key}'")
        # print(df[df.duplicated(subset=[key], keep=False)])
    else:
        print(f"{name}: '{key}' is unique.")
```

```python
# Validate primary keys
primary_keys = {
    "Customers": (customers, "customer_id"),
    "Orders": (orders, "order_id"),
    "Sellers": (sellers, "seller_id"),
    "Products": (products, "product_id"),
    "Reviews": (reviews, "review_id"),
    "Geolocations": (geolocations, "geolocation_zip_code_prefix")
}
print(f"{'Table':<20} | {'Total Rows':<10} | {'Unique IDs':<10} | {'Duplicates':<10} | {'Nulls':<5}")
print("-" * 65)
for name, (df, pk) in primary_keys.items():
    total = len(df)
    unique = df[pk].nunique()
    dupes = df[pk].duplicated().sum()
    nulls = df[pk].isnull().sum()

    status = "" if dupes == 0 and nulls == 0 else ""

    print(f"{name:<20} | {total:<10} | {unique:<10} | {dupes:<10} | {nulls:<5} {status}")
```

## Key Takeaways

1. **Data Loading**: All nine Olist datasets were successfully loaded.

2. **Data Types**: Some date columns in `orders` and `reviews` are stored as `object` and need conversion to `datetime`. Numeric columns in `products` are fine.

3. **Missing Values**: Present in `orders`, `reviews`, and `products` with varying counts.

4. **Primary Key Checks**:
   - **Unique**: `customer_id`, `order_id`, `seller_id`, `product_id`.
   - **Not unique**: `review_id` (duplicates present), `order_item_id` & `order_id` in `Payments` (expected composite keys), `geolocation_zip_code_prefix` (expected).
